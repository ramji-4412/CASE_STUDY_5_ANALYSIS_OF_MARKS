```python
# CASE STUDY 5 : Analysis of marks of students

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

df = pd.read_csv('/content/Analyisis_of_marks.csv')
max_marks = 15
df["Percentage"] = df["Marks"]/max_marks*100

above_75 = df[df["Percentage"] > 75]
between_60_75 = df[(df["Percentage"] >= 60) & (df["Percentage"] <= 75)]
below_60 = df[df["Percentage"] < 60]

print("\nStudents with > 75%:")
print(above_75)

print("\nStudents with 60% - 75%:")
print(between_60_75)

print("\nStudents with < 60%:")
print(below_60)

plt.hist(df["Percentage"], bins=range(0, 101, 5), edgecolor='black', color='blue')
plt.title("Student Percentage Distribution")
plt.xlabel("Percentage")
plt.ylabel("Number of Students")
plt.show()

plt.scatter(above_75.index, above_75["Percentage"], label=">75%", color='green',marker = "^")
plt.scatter(between_60_75.index, between_60_75["Percentage"], label="60–75%", color='orange',marker = "o")
plt.scatter(below_60.index, below_60["Percentage"], label="<60%", color='red', marker = "s")
plt.show()
```
